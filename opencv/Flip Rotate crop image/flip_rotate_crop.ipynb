```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread("./giraaffe.jpg")
```

```python
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

## Flip

```python
# 1 ----> horizontal
# 0 ----> vertical
# -1 ----> both axes

flip_horizontal = cv2.flip(image,1)
flip_vertically = cv2.flip(image,0)
flip_both_axes = cv2.flip(image,-1)

titles = ["Original image","Flip Horizontal","Flip vetically","Flip Both Axes"]
images = [image_rgb , cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_vertically,cv2.COLOR_BGR2RGB),
          cv2.cvtColor(flip_both_axes,cv2.COLOR_BGR2RGB)]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# 1 ----> horizontal
# 0 ----> vertical
# -1 ----> both axes

## also perform as 
flip_horizontal = cv2.flip(image_rgb,1)
flip_vertically = cv2.flip(image_rgb,0)
flip_both_axes = cv2.flip(image_rgb,-1)

titles = ["Original image","Flip Horizontal","Flip vetically","Flip Both Axes"]
images = [image_rgb ,flip_horizontal,flip_vertically,flip_both_axes]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## Rotate

```python
rotate_90_clockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_counterclockwise = cv2.rotate(image_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180 = cv2.rotate(image_rgb,cv2.ROTATE_180)
images = [image_rgb , rotate_90_clockwise,rotate_90_counterclockwise,rotate_180]
titles=["Original image","rotate_90_clockwise","rotate_90_counterclockwise","rotate_180"]
```

```python
plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
