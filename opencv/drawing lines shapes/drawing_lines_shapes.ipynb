```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# create a empty canvas 

height , width = 500,700
blue = (255,127,0)
canvas = np.full((height,width,3),blue,dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

```python
## Draw a line 
cv2.line(canvas,(0,0),(600,500),(0,0,0),5)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
## Draw a rectangle 

cv2.rectangle(canvas ,(100,100),(200,200),(255,255,255),-1)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

```python
# draw a circle
 
cv2.circle(canvas,(500,150),100,(255,255,255),5)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

```python
## cv2.ellipse(image_name ,(center),(major axis,minor axis),angle of ellipse , start angle ,end angle ,(color),thickness)
cv2.ellipse(canvas, (350, 350), (100, 50), 45, 0, 360, (0, 255, 0), 3)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

```python
## Draw a polygon 

# Xn, Yn, where n>2

points = np.array([[150,400],[300,400],[255,300]],np.int32)
new_reshaped_point =  points.reshape(-1,1,2)
print(new_reshaped_point) 

cv2.polylines(canvas ,[new_reshaped_point],True,(255,145,245),3)

plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))

```
